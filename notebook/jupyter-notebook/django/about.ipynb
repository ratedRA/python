import random
import os
def get_filename_ext(filepath):
	base_name = os.path.basename(filepath)
	name, ext = os.path.splitext(base_name)
	return name, ext 

def upload_image_path(instance, filename):
	newfile_name = random.randint(1,100000)
	name, ext = get_filename_ext(filename)
	final_filename  = '{newfile_name}{ext}'.format(newfile_name=newfile_name, ext = ext)
	return 'products/{newfile_name}/{final_filename}'.format(newfile_name = newfile_name,
															final_filename = final_filename)

class ProductManager(models.Manager):
	def get_by_id(self, id):
		qs = self.get_queryset().filter(id=id) # Product.objects = get_queryset()

		if qs.count()==1:
			return qs.first()
		return None
