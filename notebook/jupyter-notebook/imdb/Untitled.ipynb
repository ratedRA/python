```python
source = requests.get('https://www.imdb.com/list/ls063315922/').text

soup = BeautifulSoup(source, 'lxml')
```

```python
llist = soup.find_all('div',class_='lister-item-content')
```

```python
for elements in llist:
    name = elements.find('h3', class_='lister-item-header')
    name_fetch = name.find('a').get('href')
    add_url = 'https://www.imdb.com'+name_fetch
    new_url = requests.get(add_url).text
    soup2 = BeautifulSoup(new_url, 'lxml')
    crew = soup2.find('div', class_ = 'article')
    crew = crew.find('a').text
    year_fetch = name.find('span', class_='lister-item-year text-muted unbold').text
    year_fetch = year_fetch.replace('(',"")
    year_fetch = year_fetch.replace(')',"")
    x = elements.find('div', class_ = 'ipl-rating-widget')
    y = x.find('span', class_='ipl-rating-star__rating').text
    directors = elements.find_all('p', class_='text-muted text-small')
    for i in directors:
        for j in i.find_all('a'):
            print(j.text)
    #d_directors = d.find('a').text
    #print(d_directors)
    print(name_fetch)
    print(y)
    print(year_fetch)
    print()
    print(crew)
```
