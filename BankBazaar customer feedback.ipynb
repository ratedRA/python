```python
import bs4
import requests
import csv
```

```python
siteUrl = requests.get('https://www.bankbazaar.com/reviews.html?ck=Y%2BziX71XnZjIM9ZwEflsyDYlRL7gaN4W0xhuJSr9Iq7aMYwRm2IPACTQB2XBBtGG&rc=1').text
```

```python
siteUrl
```

```python
siteSoup = bs4.BeautifulSoup(siteUrl, 'lxml')
```

```python
csv_file = open('bankbazaarreview.csv', 'w')

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['review'])
```

```python
siteSoup.prettify()
```

```python
siteSoup
```

```python
section = siteSoup.find('section', class_= 'existing-review')
```

```python
section
```

```python
section.h2.text
```

```python
comment = section.find('strong', class_ = 'user-review-comment')
```

```python
review = section.find('div', class_ = 'text_here')
```

```python
for review in section.find_all('div', class_ = 'text_here'):
    print(review.text)
    print()
    csv_writer.writerow([review.text])
```
